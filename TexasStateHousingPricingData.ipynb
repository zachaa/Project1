```python
#import dependencies
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
from pathlib import Path
import pandas as pd
from src.get_project_data import housing_data
```

```python
#set path to CSV file
housing_df= housing_data("tx")

#show all Texas housing data
housing_df.head()


```

```python
#Group data by month

monthly_averages= housing_df.groupby(housing_df['date'])['average_listing_price'].mean()
```

```python
#show monthly averages
monthly_averages

```

```python
#find trends over texas housing prices
fig, ax = plt.subplots(figsize=(11, 6))
monthly_averages.plot()
ax.set_title("Texas Housing Prices")
ax.set_xlabel("Year")
ax.set_ylabel("Average Listing Price")
ax.vlines(x=pd.Timestamp("2020-3-01"), ymin=280000, ymax=500000, color='red', linestyle='--')
ax.title.set_size(20)
ax.xaxis.label.set_size(20)
ax.yaxis.label.set_size(20)
plt.show()
```

```python
#Group data by year
yearly_averages= housing_df.groupby(housing_df['date'].dt.year)['average_listing_price'].mean()
```
