```python
#Obtain data from the Texas State Data Center
from typing import Tuple
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st

from src.get_project_data import housing_data, locale_data
pd.options.mode.chained_assignment = None  # default='warn' #Hide SettingWithCopyWarning
```

```python
#Get the data
housing_data_Texas= housing_data()
```

```python
#Get the data for entire state of Texas, focusing on the pending listing count, that the closest to the sales per year.
housing_data_Texas = housing_data_Texas[["date", "zipcode", "zip_name", "pending_listing_count"]]  
print(housing_data_Texas.shape)    
housing_data_Texas.head()
```

```python
#Print the data
housing_data_Texas;
print(housing_data_Texas)
```

```python
#Fill the missing values with 0
housing_data_Texas ["pending_listing_count"] = housing_data_Texas[ "pending_listing_count"].fillna(0)
```

```python
#Print the data
housing_data_Texas.describe
```

```python
#Gropu the data by date and sum the pending listing count to obatin the total pending listing count for the entire state of Texas per year  
total_count_Texas = housing_data_Texas.groupby(("date"))["pending_listing_count"].sum()
housing_data_Texas.head()
```

```python

# Create a line graph to show the total pending listing count for the entire state of Texas per year
total_count_Texas.plot(kind='line', marker='o', color='blue',grid=True, figsize=(10,5))

# Set x-axis label
plt.xlabel("Year")

# Set y-axis label
plt.ylabel("Total Houses Count")
# Set title of the graph
plt.title("Texas State Pending Listing Count")


# Show the graph
plt.show()
```
